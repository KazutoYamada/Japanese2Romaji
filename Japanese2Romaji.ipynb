```python
import MeCab
```

```python
with open('Romaji1.txt', 'r') as f:
    book1 = f.read()

with open('Romaji2.txt', 'r') as f:
    book2 = f.read()

with open('Romaji3.txt', 'r') as f:
    book3 = f.read()

with open('Romaji3.txt', 'r') as f:
    book4 = f.read()

books = [book3, book2, book1, book4]
list_books = [book.split('\n') for book in books]
```

```python
txt = input('日本語のタイトルを入力してください')
#print(txt)
tagger = MeCab.Tagger('-Ochasen')
parsed_txt = tagger.parse(txt)
elements = parsed_txt.split('\n')[:-2]

wakachi_txt = []
wakachi_katakana_txt = []
list_Romaji_txt = []
for element in elements:
    wakachi_txt.append(element.split('\t')[0])
    Katakana = element.split('\t')[1]
    Romaji = Katakana
    for list_book in list_books:
        for moji in list_book:
            kana, roman = moji.split('\t')
            Romaji = Romaji.replace(kana, roman)
    wakachi_katakana_txt.append(Katakana)
    list_Romaji_txt.append(Romaji)

    #if no == 0:
        #romaji = romaji.capitalize()

Romaji_txt = ' '.join(list_Romaji_txt)
character_list_Romaji_txt = list(Romaji_txt)
i = 0
for romaji in character_list_Romaji_txt:
    if romaji == 'ッ':
        character_list_Romaji_txt[i] = character_list_Romaji_txt[i+1]
    i += 1
Romaji_txt = ''.join(character_list_Romaji_txt).capitalize()
print(*wakachi_txt)
print(Romaji_txt.replace('ou','oo'))
```
